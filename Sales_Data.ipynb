```python
q1=sales["QUANTITY"].quantile(0.25)
q3=sales["QUANTITY"].quantile(0.75)
iqr=q3-q1
```

```python
upper_limit=q3 +1.5*iqr
lower_limit=q1-1.5*iqr
```

```python
#capping
sales["QUANTITY"]=np.where(sales["QUANTITY"]>upper_limit,
                           upper_limit,
                           np.where(sales["QUANTITY"]<lower_limit,
                                    lower_limit,
                                    sales["QUANTITY"]))
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.linear_model import LinearRegression
```

```python
step1=ColumnTransformer(transformers=[("col_tnf",OneHotEncoder(sparse_output=False,handle_unknown="ignore",drop="first"),["MONTH","PRODUCT_CODE","PRODUCT_DESCRIPTION"])]
                        ,remainder="passthrough")
step2=XGBRegressor(max_depth=3,n_estimators=1500,learning_rate=0.2)
pipe=Pipeline([("step1",step1),("step2",step2)])
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
print("R2_Score",r2_score(y_test,y_pred))
print("MAE",mean_absolute_error(y_test,y_pred))
```
